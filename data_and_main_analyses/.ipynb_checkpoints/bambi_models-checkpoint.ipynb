# Model-Based and Model-Free Effects

$\text{choice}_{t} =\overbrace{\text{intercept}}^\text{Action Perseveration} + \overbrace{\text{rwd}_{t-1}+\text{pun}_{t-1}}^\text{Goal Perseveration} +\text{goal}_{t-1}+\overbrace{\text{rwd}_{t-1}X\text{goal}_{t}+\text{pun}_{t-1}X\text{goal}_{t}}^\text{Model-Based}+\overbrace{\text{rwd}_{chosen,t-1}X\text{goal}_{t-1}+\text{pun}_{chosen,t-1}X\text{goal}_{t-1}}^\text{Model-Free}$.


- Features: 1 if present, -1 if absent.
- Goals: 1 if present, -1 if absent.
- Interactions: 1 if goal & feature present, 0 if goal absent, -1 if goal present but feature absent.




```python
import bambi as bmb
import pandas as pd
df1='switches_bambi_full_cleaned.csv'
```

```python
dfm = bmb.Model("switching ~ r+p+g+mbrew+mbpun+mfrew+mfpun+(r|subject)+(p|subject)+(g|subject)+(mbrew|subject)+(mbpun|subject)+(mfrew|subject)+(mfpun|subject)",data=df1,family="bernoulli")
model_fitted = dfm.fit(draws=2000, target_accept=0.85)
```

```python
import arviz as az
import bambi as bmb
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
dfm.plot_priors()
az.summary(model_fitted,var_names=['r','p','r|subject_sigma','mfrew|subject_sigma','mbpun'])
plt.savefig('bambi_priors.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python

az.plot_posterior(
    model_fitted,
    var_names=['r','p'],
    hdi_prob=0.95,
    textsize=14
    
);
plt.show()
```

```python
az.plot_posterior(
    model_fitted,
    var_names=['mfrew|subject_sigma','mfrew'],
    hdi_prob=0.95,
    ref_val=0
);
# plt.savefig('mf_posteriors.png', bbox_inches='tight',  dpi=300)
plt.show()
```

```python
az.plot_posterior(
    model_fitted,
    var_names=['mbrew','mbpun'],
    ref_val=0,
);
plt.savefig('mb_posteriors.png', bbox_inches='tight',  dpi=300)
plt.show()
```

```python
## import matplotlib.pyplot as plt
import seaborn as sns
from bayesian_bootstrap.bootstrap import mean, highest_density_interval
import pandas as pd
import matplotlib.pyplot as plt

import numpy as np

#Extract data
gprew = model_fitted.posterior['r'].values.tolist()

gprew = sum(gprew, [])
np.save('gprews_empirical2.npy',gprew)

gppun = model_fitted.posterior['p'].values.tolist()
gppun = sum(gppun, [])
np.save('gppuns_empirical2.npy',gppun)

mbrew = model_fitted.posterior['mbrew'].values.tolist()
mbrew = sum(mbrew, [])
np.save('mbrews_empirical2.npy',mbrew)


mbpun = model_fitted.posterior['mbpun'].values.tolist()
mbpun = sum(mbpun, [])
np.save('mbpuns_empirical2.npy',mbpun)

mfrew = model_fitted.posterior['mfrew'].values.tolist()
mfrew = sum(mfrew, [])
np.save('mfrews_empirical2.npy',mfrew)

mfpun = model_fitted.posterior['mfpun'].values.tolist()
mfpun = sum(mfpun, [])
np.save('mfpuns_empirical2.npy',mfpun)


# CONTRASTS
gpdiff=[np.abs(gprew[i])-gppun[i] for i in range(len(gprew))]
mbdiff=[np.abs(mbrew[i])-mbpun[i] for i in range(len(mbrew))]
mfdiff=[np.abs(mfrew[i])-mfpun[i] for i in range(len(mbrew))]

gp_maineffect=[np.abs(gprew[i])+gppun[i]/2.0 for i in range(len(gprew))]
mb_maineffect=[np.abs(mbrew[i])+mbpun[i]/2.0 for i in range(len(mbpun))]
mf_maineffect=[np.abs(mfrew[i])+mfpun[i]/2.0 for i in range(len(mbpun))]



#Compute HDIs
l_mbr, r_mbr = highest_density_interval(mbrew,alpha=0.05)
l_mbp, r_mbp = highest_density_interval(mbpun,alpha=0.05)
l_gpr, r_gpr = highest_density_interval(gprew,alpha=0.05)
l_gpp, r_gpp = highest_density_interval(gppun,alpha=0.05)
l_mfr, r_mfr = highest_density_interval(mfrew,alpha=0.05)
l_mfp, r_mfp = highest_density_interval(mfpun,alpha=0.05)
l_dmb, r_dmb = highest_density_interval(mbdiff,alpha=0.05)
l_dgp, r_dgp = highest_density_interval(gpdiff,alpha=0.05)
l_dmf, r_dmf = highest_density_interval(mfdiff,alpha=0.05)

l_mb, r_mb = highest_density_interval(mb_maineffect,alpha=0.05)
l_gp, r_gp = highest_density_interval(gp_maineffect,alpha=0.05)
l_mf, r_mf = highest_density_interval(mf_maineffect,alpha=0.05)

rope=0.002
```
