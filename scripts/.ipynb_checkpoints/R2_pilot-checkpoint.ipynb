# Task features impacting reward benefits of attending to features
- timeseries of reward function doesn't appear to impact 


- feature trajectories do
    - if one feature has a greater disparity in feature probabilities, impacts learning
    - if a feature in both states hovers around 50%, there's not much to learn
    - created a new criterion that the difference between trajectories for a feature across states should be relatively the same for both features (difference of a difference)
    
        - $abs(abs(\bar{f_{1,1}}-\bar{f{1,2}})-abs(\bar{f_{2,1}}-\bar{f_{2,2}})) <0.05$
        
        
- with this new criterion, 14/20 trajectories yielded benefit for environment-matched feature
- held learning rate constant (this might interact with feature trajectory depending on volatility)
    
    
- Will counterbalance the feature trajectory participants experience: there will now be 4 trajectories
    - for safe first, 2, and for danger first, 2
    
    
- In terms of points earned, only 2/16 subjects performed around levels expected by agents that purely guess

```python
import pandas as pd
sr=pd.read_csv('run2a2_data.csv')
sr=sr[['Q_name','responses','subjectID']]
valid_columns=['anx_aro','worry']
sr1=sr[sr.Q_name.isin(valid_columns)]
print(len(sr1))
rs_items_worry=[1,3,8,10,11]
rs_dict={1:5,2:5,4:2,5:1,3:3}
sr1=sr1.reset_index(drop=True)
```

```python
import json
from scipy.stats import kendalltau as corr
sr1.responses

# print(sr.subjectID.unique())
    
worry_scores=[]
masq_scores=[]

for entry in sr1.responses:
    x=json.loads(entry)
    if len(x)==16:
        sum_worry=0
        for key in x.keys():
            if int(key[1:])+1 in rs_items_worry:
                sum_worry+=rs_dict[int(x[key])+1]
            else:
                sum_worry+=int(x[key])+1
        worry_scores.append(sum_worry)
    else:
        masq_sum=0
        for key in x.keys():
            masq_sum+=int(x[key])+1
        masq_scores.append(masq_sum)
        
print(worry_scores)
            
print(masq_scores)
r,p=corr(masq_scores,worry_scores)
print('correl masq and worry: {}'.format(r))
safe_betas=[7.447730071153159, 3.571634141019478, 4.962151288081429, 3.3580538377503055, 4.104699662774458, 4.091363854396553, 8.75960876580978, 3.549672431385079, 9.844030449358922, 9.108967421429943, 2.2348403876269147, 4.373370822361022, 2.3040948171327322, 3.112652485298704, 3.9653040001289863, 2.692773841234405]
pred_betas=[1.3760634558542595, 2.7120182852951578, 4.206658361881413, 3.1780365631438934, 2.5295003190173198, 3.339680800395735, 12.690539595481582, 3.597966046145405, 17.94574540408061, 8.607339057296285, 5.36573179463215, 0.4691935475476812, 1.3608968624669693, 1.8970105509221735, 3.8259096213886017, 2.437764980007427]
r,p=corr(safe_betas,pred_betas)
print('correl safe and pred: {}'.format(r))
r,p=corr(worry_scores,pred_betas)
print('correl worry and pred: {}'.format(r))
r,p=corr(worry_scores,safe_betas)
print('correl worry and safe: {}'.format(r))
r,p=corr(masq_scores,pred_betas)
print('correl arousal and pred: {}'.format(r))
r,p=corr(masq_scores,safe_betas)
print('correl arousal and safe: {}'.format(r))
```

# Safe First Order : Safe Environment

```python
import numpy as np
from scipy.stats import spearmanr as corrs
#Safe Env First and Safe condition
switching=np.load('Switching_safefirst_safe_full.npy')
switching=switching.astype(np.int64)
predictors=np.load('Predictors_safefirst_safe_full.npy')
drop_rows=[2,3,6,8,9,11,12,13,14,15,16,17]
# initial_trials=np.arange(10)
# predictors=np.delete(predictors,initial_trials,axis=1)

predictors=np.delete(predictors,drop_rows,axis=2)
# switching=np.delete(switching,drop_rows,axis=2)

print(predictors.shape)

# print(predictors[1,:,10][0:140])
# print(predictors[1,:,9][0:140])

# switching=np.delete(switching,initial_trials,axis=1)
print(switching.shape)
# r,p=corrs(predictors[1,:,2],predictors[1,:,5])
# print(r)
```

```python
from pystan import StanModel
#     int<lower=0,upper=1> SafeRF[NxSubjects, NxChoices];
#     int<lower=0,upper=1> PredRF[NxSubjects, NxChoices];
#     int<lower=0,upper=1> lastO_Pred[NxSubjects, NxChoices];
#     int<lower=0,upper=1> lastO_Safe[NxSubjects, NxChoices];
#     int<lower=0,upper=1> InteractionPred[NxSubjects, NxChoices];
#     int<lower=0,upper=1> InteractionSafe[NxSubjects, NxChoices];
model_input='''
data {

    int<lower=1> NxChoices;
    int<lower=1> NxSubjects;
    int<lower=1> NxPredictors;
    int<lower=0,upper=1> Choice[NxSubjects,NxChoices];
    matrix[NxChoices,NxPredictors] Predictors[NxSubjects];
}
parameters {

  // Group level
  real mu[NxPredictors];
  real<lower=0,upper=2> sigma;
  
  //Individual Level
  vector[NxPredictors] beta[NxSubjects];

}


model {    
    sigma ~ uniform(0,2);
    for (d in 1:NxPredictors) {
        mu[d] ~ normal(0, 10);
        for (l in 1:NxSubjects){
          beta[l,d] ~ normal(mu[d],sigma);
          }
        }
    for (k in 1:NxSubjects){
        
        
        // Neutral condition choices
        Choice[k,:] ~ bernoulli(inv_logit(Predictors[k,:]*beta[k,:]));
     
          }    
      
    }
    
'''
data_input = {
              'Choice':switching,#choice data
              'NxChoices':80,
              'NxSubjects':8,
              'NxPredictors':6,
              'Predictors':predictors
}

controls={'adapt_delta':0.9,
         'max_treedepth':18}

model_fit_Anova = StanModel(model_code=model_input)
fit_model_fullAnova_nocontrol = model_fit_Anova.sampling(data=data_input,iter=2000,control=controls)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from bayesian_bootstrap.bootstrap import mean, highest_density_interval
import pandas as pd

# import pickle
# with open("ANOVA_r2_model_fit.pkl", "wb") as f:
#     pickle.dump({'model' : model_input, 'fit' : fit}, f, protocol=-1)


fit_df_2 = fit_model_fullAnova_nocontrol.to_dataframe()
#fit_df_2.to_csv('Bayesian_r2_RandomEffects.csv')
#fit_df=pd.read_csv('Bayesian_rmANOVA_output_r2_removedOutlier_groupPriorSDs_noInteraction_covaryingValueatChoice.csv')
#Extract data
chain_idx = fit_df_2['draw']
predator_learning = fit_df_2['mu[{}]'.format(5)]
reward_learning = fit_df_2['mu[{}]'.format(6)]


# fif=fit_df_2['bConditions[{}]'.format(9)]
#Compute Contrasts


#Compute HDIs
l_ct, r_ct = highest_density_interval(predator_learning)
l_cp, r_cp = highest_density_interval(reward_learning)


rope=0.002


sns.set(style="white",
        palette="muted",
        color_codes=True)

f, axs = plt.subplots(2, 1, figsize=(8, 5),sharex=False)

ax0=sns.distplot(predator_learning, hist=False,kde_kws={"shade": True},color="k", ax=axs[0])
ax0.set(title='Learning for predator features',xlabel='')

axs[0].plot([l_ct, r_ct],[0,0],linewidth=8.0,label='95% HDI',marker='o',color='y')
axs[0].plot([-rope, rope],[0,0],linewidth=8.0,label='ROPE',marker='o',color='black')


ax1=sns.distplot(reward_learning, hist=False,kde_kws={"shade": True},color="g", ax=axs[1])
ax1.set(title='Learning for reward features',xlabel='')

axs[1].plot([l_cp, r_cp],[0,0],linewidth=8.0,label='95% HDI',marker='o',color='y')
axs[1].plot([-rope, rope],[0,0],linewidth=8.0,label='ROPE',marker='o',color='black')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import os
initial_trials=np.arange(20)

#Safe Env First and Safe condition
switching1=np.load('Switching_safefirst_safe_full.npy')
switching1=switching1.astype(np.int64)
# switching1=np.delete(switching1,initial_trials,axis=1)

switching2=np.load('Switching_safefirst_danger_full.npy')
switching2=switching2.astype(np.int64)
# switching2=np.delete(switching2,initial_trials,axis=1)

switching4=np.load('Switching_dangerfirst_safe_full.npy')
switching4=switching4.astype(np.int64)
# switching4=np.delete(switching4,initial_trials,axis=1)

switching3=np.load('Switching_dangerfirst_danger_full.npy')
switching3=switching3.astype(np.int64)
# switching3=np.delete(switching3,initial_trials,axis=1)



predictors1=np.load('Predictors_safefirst_safe_full.npy')
# predictors1=np.delete(predictors1,initial_trials,axis=1)

predictors2=np.load('Predictors_safefirst_danger_full.npy')
# predictors2=np.delete(predictors2,initial_trials,axis=1)

predictors3=np.load('Predictors_dangerfirst_danger_full.npy')
# predictors3=np.delete(predictors3,initial_trials,axis=1)

predictors4=np.load('Predictors_dangerfirst_safe_full.npy')
# predictors4=np.delete(predictors4,initial_trials,axis=1)





p1=np.concatenate((predictors1,predictors2),axis=1)
p2=np.concatenate((predictors4,predictors3),axis=1)

p=np.concatenate((p1,p2),axis=0)

# drop_cols=[3,4,5]
# p=np.delete(p,drop_cols,axis=2)
s1=np.concatenate((switching1,switching2),axis=1)
print('mean of safe first: {}'.format(np.mean(s1)))

s2=np.concatenate((switching3,switching4),axis=1)
print('mean of danger first: {}'.format(np.mean(s2)))



s=np.concatenate((s1,s2),axis=0)
print('grand mean: {}'.format(np.mean(s)))

# predictors=np.concatenate((predictors1,predictors2),axis=0)

print(p.shape)
print(s.shape)
```

```python
from pystan import StanModel

model_input='''
data {

    int<lower=1> NxChoices;
    int<lower=1> NxSubjects;
    int<lower=1> NxPredictors;
    int<lower=0,upper=1> Choice[NxSubjects,NxChoices];
    matrix[NxChoices,NxPredictors] Predictors[NxSubjects];

}
parameters {

  // Group level
  real mu[NxPredictors-5];
  real aOrder;
  real Order_Pred;
  real Order_Pred_Env;
  real Order_Reward;
  real Order_Reward_Env;
  
  //Individual Level
  vector[NxPredictors-5] beta1[NxSubjects];

}
transformed parameters{
vector[NxPredictors] beta_f[NxSubjects];

//for fixed effects 
for (d in 1:NxPredictors) {
for (l in 1:NxSubjects){
if (d==14){
beta_f[l,d]=aOrder;
}
else if (d==15){
beta_f[l,d]=Order_Pred;
}
else if (d==16){
beta_f[l,d]=Order_Reward;
}
else if (d==17){
beta_f[l,d]=Order_Pred_Env;
}
else if (d==18){
beta_f[l,d]=Order_Reward_Env;
}
else {
beta_f[l,d]=beta1[l,d];
}
}
}
}
model {    
    aOrder ~ normal(0,1);
    Order_Pred ~ normal(0,1);
    Order_Pred_Env ~ normal(0,1); 
    Order_Reward ~ normal(0,1);
    Order_Reward_Env ~ normal(0,1);
    
    for (d in 1:NxPredictors-5) {
        
        mu[d] ~ normal(0,1);
        
        for (l in 1:NxSubjects){
          beta1[l,d] ~ normal(mu[d],1);
          }
        }

        
  
    for (k in 1:NxSubjects){
        Choice[k,:] ~ bernoulli(inv_logit(Predictors[k,:]*beta_f[k,:]));
        }    
    }
    
'''
data_input = {
              'Choice':s,#choice data
              'NxChoices':160,
              'NxSubjects':16,
              'NxPredictors':18,
              'Predictors':p,

    
}

controls={'adapt_delta':0.9,
         'max_treedepth':18}

model_fit_Anova = StanModel(model_code=model_input)
full_model_all = model_fit_Anova.sampling(data=data_input,iter=2000,control=controls)
```

```python
from pystan import StanModel
#     int<lower=0,upper=1> SafeRF[NxSubjects, NxChoices];
#     int<lower=0,upper=1> PredRF[NxSubjects, NxChoices];
#     int<lower=0,upper=1> lastO_Pred[NxSubjects, NxChoices];
#     int<lower=0,upper=1> lastO_Safe[NxSubjects, NxChoices];
#     int<lower=0,upper=1> InteractionPred[NxSubjects, NxChoices];
#     int<lower=0,upper=1> InteractionSafe[NxSubjects, NxChoices];
model_input='''
data {

    int<lower=1> NxChoices;
    int<lower=1> NxSubjects;
    int<lower=1> NxPredictors;
    int<lower=0,upper=1> Choice[NxSubjects,NxChoices];
    matrix[NxChoices,NxPredictors] Predictors[NxSubjects];
}

parameters {

  real mu[NxPredictors];
  real<lower=0,upper=2> sigma;
  
  //Individual Level
  vector[NxPredictors] beta[NxSubjects];
  

}


model {    
    sigma ~ uniform(0,2);
    for (d in 1:NxPredictors) {
        mu[d] ~ normal(0, 10);
        for (l in 1:NxSubjects){
          beta[l,d] ~ normal(mu[d],sigma);
          }
        }
    for (k in 1:NxSubjects){
        
        
        // Neutral condition choices
        Choice[k,:] ~ bernoulli(inv_logit(Predictors[k,:]*beta[k,:]));
     
          }    
      
    }
    
'''
data_input = {
              'Choice':s,#choice data
              'NxChoices':160,
              'NxSubjects':16,
              'NxPredictors':6,
              'Predictors':p
}

controls={'adapt_delta':0.9,
         'max_treedepth':18}

model_fit_Anova = StanModel(model_code=model_input)
fit_model_full = model_fit_Anova.sampling(data=data_input,iter=2000,control=controls)
```

# Reduced model: Evidence for learning w/o taking into account counterfactuals

F1= reward feature
F2= medium feature
F3= predator feature

$\beta$ 's are main effects, and $\gamma$'s are interaction effects.

$switch \sim Bernoulli(\theta_{i})$

$\theta = logistic(intercept+\beta_{f1}+\beta_{f2}+\beta_{rewRF}+\beta_{predRF}+\gamma_{f1,rewRF}+\gamma_{f2,midRF})$

Each parameter drawn from group level prior, whose mean is estimated: $\mathcal{N} \sim (\mu,1)$

Means all drawn from hierarchical unit normal prior.

Learning is captured by interaction terms. If I encounter a reward feature in a given state, and next trial, I face the goal to seek reward, I should stay more with my previous action.

Because model is predicting switching, negative coefficients represent greater staying behaviour.

## Add a model where interaction terms defining feature learning depends on counterfactual feedback: 

Below $\beta_{otherRew}$ is 1 if no other states had a reward feature on last trial.

$\psi$ represent 3-way interactions, wherein the interactions defining feature learning may or may not depend on counterfactual feedback


$\theta = logistic(intercept+\beta_{f1}+\beta_{f2}+\beta_{f3}+\beta_{rewRF}+\beta_{predRF}+\beta_{otherRew}+\beta_{otherPred}+\gamma_{f1,rewRF}+\gamma_{f2,midRF}+\gamma_{f3,predRF}+\psi_{f1,rewRF,otherRew}+\psi_{f3,predRF,otherPred})$

```python
##### import numpy as np
import os

#Safe Env First and Safe condition
switching1=np.load('Switching_safefirst_safe_full.npy')
switching1=switching1.astype(np.int64)

switching2=np.load('Switching_safefirst_danger_full.npy')
switching2=switching2.astype(np.int64)
switching4=np.load('Switching_dangerfirst_safe_full.npy')
switching4=switching4.astype(np.int64)
switching3=np.load('Switching_dangerfirst_danger_full.npy')
switching3=switching3.astype(np.int64)

predictors1=np.load('Predictors_safefirst_safe_full.npy')
predictors2=np.load('Predictors_safefirst_danger_full.npy')
predictors3=np.load('Predictors_dangerfirst_danger_full.npy')
predictors4=np.load('Predictors_dangerfirst_safe_full.npy')




p1=np.concatenate((predictors1,predictors2),axis=1)
p2=np.concatenate((predictors4,predictors3),axis=1)

p=np.concatenate((p1,p2),axis=0)
print(p.shape)
drop_rows=[6,9,11,13,14,15,16,17]
p=np.delete(p,drop_rows,axis=2)
s1=np.concatenate((switching1,switching2),axis=1)
print('mean of safe first: {}'.format(np.mean(s1)))

s2=np.concatenate((switching3,switching4),axis=1)
print('mean of danger first: {}'.format(np.mean(s2)))



s=np.concatenate((s1,s2),axis=0)
print('grand mean: {}'.format(np.mean(s)))

# predictors=np.concatenate((predictors1,predictors2),axis=0)

print(p.shape)
print(s.shape)
```

```python
from pystan import StanModel

model_input='''
data {

    int<lower=1> NxChoices;
    int<lower=1> NxSubjects;
    int<lower=1> NxPredictors;
    int<lower=0,upper=1> Choice[NxSubjects,NxChoices];
    matrix[NxChoices,NxPredictors] Predictors[NxSubjects];

}
parameters {

  // Group level
  real mu[NxPredictors];

  
  //Individual Level
  vector[NxPredictors] beta1[NxSubjects];

}

model {    
    
    for (d in 1:NxPredictors) {
        
        mu[d] ~ normal(0,1);
        
        for (l in 1:NxSubjects){
          beta1[l,d] ~ normal(mu[d],1);
          }
        }

        
  
    for (k in 1:NxSubjects){
        Choice[k,:] ~ bernoulli(inv_logit(Predictors[k,:]*beta1[k,:]));
        }    
    }
    
'''
data_input = {
              'Choice':s,#choice data
              'NxChoices':160,
              'NxSubjects':16,
              'NxPredictors':10,
              'Predictors':p,

    
}

controls={'adapt_delta':0.9,
         'max_treedepth':18}

model_fit_Anova = StanModel(model_code=model_input)
cf_model = model_fit_Anova.sampling(data=data_input,iter=2000,control=controls)
```
