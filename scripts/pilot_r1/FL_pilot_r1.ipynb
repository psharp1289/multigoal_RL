```python
from pystan import StanModel
#     int<lower=0,upper=1> SafeRF[NxSubjects, NxChoices];
#     int<lower=0,upper=1> PredRF[NxSubjects, NxChoices];
#     int<lower=0,upper=1> lastO_Pred[NxSubjects, NxChoices];
#     int<lower=0,upper=1> lastO_Safe[NxSubjects, NxChoices];
#     int<lower=0,upper=1> InteractionPred[NxSubjects, NxChoices];
#     int<lower=0,upper=1> InteractionSafe[NxSubjects, NxChoices];
model_input='''
data {

    int<lower=1> NxChoices;
    int<lower=1> NxSubjects;
    int<lower=1> NxPredictors;
    int<lower=0,upper=1> Choice[NxSubjects,NxChoices];
    matrix[NxChoices,NxPredictors] Predictors[NxSubjects];
}
parameters {

  // Group level
  real mu[NxPredictors];
  real<lower=0,upper=2> sigma;
  
  //Individual Level
  vector[NxPredictors] beta[NxSubjects];

}


model {    
    sigma ~ uniform(0,2);
    for (d in 1:NxPredictors) {
        mu[d] ~ normal(0, 10);
        for (l in 1:NxSubjects){
          beta[l,d] ~ normal(mu[d],sigma);
          }
        }
    for (k in 1:NxSubjects){
        
        
        // Neutral condition choices
        Choice[k,:] ~ bernoulli(inv_logit(Predictors[k,:]*beta[k,:]));
     
          }    
      
    }
    
'''
data_input = {
              'Choice':switching,#choice data
              'NxChoices':140,
              'NxSubjects':8,
              'NxPredictors':7,
              'Predictors':predictors
}

controls={'adapt_delta':0.9,
         'max_treedepth':18}

model_fit_Anova = StanModel(model_code=model_input)
fit_model_fullAnova_nocontrol = model_fit_Anova.sampling(data=data_input,iter=2000,control=controls)
```

```python
from pystan import StanModel
#     int<lower=0,upper=1> SafeRF[NxSubjects, NxChoices];
#     int<lower=0,upper=1> PredRF[NxSubjects, NxChoices];
#     int<lower=0,upper=1> lastO_Pred[NxSubjects, NxChoices];
#     int<lower=0,upper=1> lastO_Safe[NxSubjects, NxChoices];
#     int<lower=0,upper=1> InteractionPred[NxSubjects, NxChoices];
#     int<lower=0,upper=1> InteractionSafe[NxSubjects, NxChoices];
model_input='''
data {

    int<lower=1> NxChoices;
    int<lower=1> NxSubjects;
    int<lower=1> NxPredictors;
    int<lower=0,upper=1> Choice[NxSubjects,NxChoices];
    int<lower=0,upper=1> Choice2[NxSubjects,NxChoices];
    matrix[NxChoices,NxPredictors] Predictors[NxSubjects];
    matrix[NxChoices,NxPredictors] Predictors2[NxSubjects];
}
parameters {

  // Group level
  real mu[NxPredictors+2];
  real<lower=0,upper=2> sigma;
  
  //Individual Level
  vector[NxPredictors] beta1[NxSubjects];
  vector[2] beta_temp[NxSubjects];

}

transformed parameters{
vector[NxPredictors] beta2[NxSubjects];

for (d in 1:NxPredictors) {
        for (l in 1:NxSubjects){
          if (d==4){
          beta2[l,d]=beta_temp[l,1];
          }
          else if (d==7){
          beta2[l,d]=beta_temp[l,2];
          }
          else{
          beta2[l,d]=beta1[l,d];
          }
        }
        
        }
        

}
model {    
    sigma ~ uniform(0,2);
    for (d in 1:NxPredictors) {
        mu[d] ~ normal(0, 10);
        for (l in 1:NxSubjects){
          beta1[l,d] ~ normal(mu[d],sigma);
          }
        }
    mu[NxPredictors+1] ~ normal(0,10);
    mu[NxPredictors+2] ~ normal(0,10);
    for (l in 1:NxSubjects){
        beta_temp[l,1] ~ normal(mu[NxPredictors+1],sigma);
        beta_temp[l,2] ~ normal(mu[NxPredictors+2],sigma);      
    
    }
    for (k in 1:NxSubjects){
        
        
        // Neutral condition choices
        Choice[k,:] ~ bernoulli(inv_logit(Predictors[k,:]*beta1[k,:]));
        Choice2[k,:] ~ bernoulli(inv_logit(Predictors2[k,:]*beta2[k,:]));
     
          }    
      
    }
    
'''
data_input = {
              'Choice':switching1,#choice data
              'Choice2':switching2,
              'NxChoices':140,
              'NxSubjects':7,
              'NxPredictors':7,
              'Predictors':predictors1,
              'Predictors2':predictors2
}

controls={'adapt_delta':0.9,
         'max_treedepth':18}

model_fit_Anova = StanModel(model_code=model_input)
fit_model_combined_d1 = model_fit_Anova.sampling(data=data_input,iter=2000,control=controls)
```

```python
import numpy as np
#Safe Env First and Safe condition
switching1=np.load('Switching_dangerfirst_danger_cleaned.npy')
switching1=switching1.astype(np.int64)
predictors1=np.load('Predictors_dangerfirst_danger_cleaned.npy')
switching2=np.load('Switching_safefirst_danger_cleaned.npy')
switching2=switching2.astype(np.int64)
predictors2=np.load('Predictors_safefirst_danger_cleaned.npy')


switching=np.concatenate((switching1,switching2),axis=0)

predictors=np.concatenate((predictors1,predictors2),axis=0)

print(predictors.shape)
print(predictors[:,:,5].shape)
print(switching.shape)
```

```python
from pystan import StanModel
#     int<lower=0,upper=1> SafeRF[NxSubjects, NxChoices];
#     int<lower=0,upper=1> PredRF[NxSubjects, NxChoices];
#     int<lower=0,upper=1> lastO_Pred[NxSubjects, NxChoices];
#     int<lower=0,upper=1> lastO_Safe[NxSubjects, NxChoices];
#     int<lower=0,upper=1> InteractionPred[NxSubjects, NxChoices];
#     int<lower=0,upper=1> InteractionSafe[NxSubjects, NxChoices];
model_input='''
data {

    int<lower=1> NxChoices;
    int<lower=1> NxSubjects;
    int<lower=1> NxPredictors;
    int<lower=0,upper=1> Choice[NxSubjects,NxChoices];
    matrix[NxChoices,NxPredictors] Predictors[NxSubjects];
}
parameters {

  // Group level
  real mu[NxPredictors];
  real<lower=0,upper=2> sigma;
  
  //Individual Level
  vector[NxPredictors] beta[NxSubjects];

}


model {    
    sigma ~ uniform(0,2);
    for (d in 1:NxPredictors) {
        mu[d] ~ normal(0, 10);
        for (l in 1:NxSubjects){
          beta[l,d] ~ normal(mu[d],sigma);
          }
        }
    for (k in 1:NxSubjects){
        
        
        // Neutral condition choices
        Choice[k,:] ~ bernoulli(inv_logit(Predictors[k,:]*beta[k,:]));
     
          }    
      
    }
    
'''
data_input = {
              'Choice':switching,#choice data
              'NxChoices':140,
              'NxSubjects':15,
              'NxPredictors':7,
              'Predictors':predictors
}

controls={'adapt_delta':0.9,
         'max_treedepth':18}

model_fit_Anova = StanModel(model_code=model_input)
fit_model_combined_dangerEnv = model_fit_Anova.sampling(data=data_input,iter=2000,control=controls)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from bayesian_bootstrap.bootstrap import mean, highest_density_interval
import pandas as pd

# import pickle
# with open("ANOVA_r2_model_fit.pkl", "wb") as f:
#     pickle.dump({'model' : model_input, 'fit' : fit}, f, protocol=-1)


fit_df_2 = fit_model_combined_dangerEnv.to_dataframe()
#fit_df_2.to_csv('Bayesian_r2_RandomEffects.csv')
#fit_df=pd.read_csv('Bayesian_rmANOVA_output_r2_removedOutlier_groupPriorSDs_noInteraction_covaryingValueatChoice.csv')
#Extract data
chain_idx = fit_df_2['draw']
predator_learning = fit_df_2['mu[{}]'.format(4)]
reward_learning = fit_df_2['mu[{}]'.format(7)]


# fif=fit_df_2['bConditions[{}]'.format(9)]
#Compute Contrasts


#Compute HDIs
l_ct, r_ct = highest_density_interval(predator_learning)
l_cp, r_cp = highest_density_interval(reward_learning)


rope=0.002


sns.set(style="white",
        palette="muted",
        color_codes=True)

f, axs = plt.subplots(2, 1, figsize=(8, 5),sharex=True)

ax0=sns.distplot(predator_learning, hist=False,kde_kws={"shade": True},color="k", ax=axs[0])
ax0.set(title='Learning for predator features',xlabel='')

axs[0].plot([l_ct, r_ct],[0,0],linewidth=8.0,label='95% HDI',marker='o',color='y')
axs[0].plot([-rope, rope],[0,0],linewidth=8.0,label='ROPE',marker='o',color='black')


ax1=sns.distplot(reward_learning, hist=False,kde_kws={"shade": True},color="g", ax=axs[1])
ax1.set(title='Learning for reward features',xlabel='')

axs[1].plot([l_cp, r_cp],[0,0],linewidth=8.0,label='95% HDI',marker='o',color='y')
axs[1].plot([-rope, rope],[0,0],linewidth=8.0,label='ROPE',marker='o',color='black')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
#Safe Env First and Safe condition
switching1=np.load('Switching_dangerfirst_danger_cleaned.npy')
switching1=switching1.astype(np.int64)
predictors1=np.load('Predictors_dangerfirst_danger_cleaned.npy')
switching2=np.load('Switching_safefirst_danger_cleaned.npy')
switching2=switching2.astype(np.int64)
predictors2=np.load('Predictors_safefirst_danger_cleaned.npy')


import numpy as np
#Safe Env First and Safe condition
switching3=np.load('Switching_dangerfirst_safe_cleaned.npy')
switching3=switching1.astype(np.int64)
predictors3=np.load('Predictors_dangerfirst_safe_cleaned.npy')
switching4=np.load('Switching_safefirst_safe_cleaned.npy')
switching4=switching2.astype(np.int64)
predictors4=np.load('Predictors_safefirst_safe_cleaned.npy')

switching=np.concatenate((switching1,switching4),axis=0)
switching1a=np.concatenate((switching3,switching2),axis=0)
predictors=np.concatenate((predictors1,predictors4),axis=0)
predictors1a=np.concatenate((predictors3,predictors2),axis=0)

print(predictors1a.shape)
print(switching.shape)


# drop=[14]
# all_predictors=np.delete(all_predictors,drop,axis=0)
# all_switching=np.delete(all_switching,drop,axis=0)

```

```python
from pystan import StanModel
#     int<lower=0,upper=1> SafeRF[NxSubjects, NxChoices];
#     int<lower=0,upper=1> PredRF[NxSubjects, NxChoices];
#     int<lower=0,upper=1> lastO_Pred[NxSubjects, NxChoices];
#     int<lower=0,upper=1> lastO_Safe[NxSubjects, NxChoices];
#     int<lower=0,upper=1> InteractionPred[NxSubjects, NxChoices];
#     int<lower=0,upper=1> InteractionSafe[NxSubjects, NxChoices];
model_input='''
data {

    int<lower=1> NxChoices;
    int<lower=1> NxSubjects;
    int<lower=1> NxPredictors;
    int<lower=0,upper=1> ChoiceA[NxSubjects,NxChoices];
    int<lower=0,upper=1> ChoiceB[NxSubjects,NxChoices];
    matrix[NxChoices,NxPredictors] PredictorsA[NxSubjects];
    matrix[NxChoices,NxPredictors] PredictorsB[NxSubjects];
}


parameters {

  // Group level
  real mu[NxPredictors];
  real<lower=0,upper=2> sigma;
  vector[2] aOrder;
  
  //Individual Level
  vector[NxPredictors] beta[NxSubjects];
  

}

transformed parameters {
vector[NxPredictors] beta1[NxSubjects];
vector[NxPredictors] beta2[NxSubjects];

for (i in 1:NxSubjects){
for (h in 1:NxPredictors){
if (h==4){
beta1[i,h]=beta[i,h]+aOrder[1];
beta2[i,h]=beta[i,h]+aOrder[2];
}
else if (h==7){
beta1[i,h]=beta[i,h]+aOrder[1];
beta2[i,h]=beta[i,h]+aOrder[2];

}
else{
beta1[i,h]=beta[i,h];
beta2[i,h]=beta[i,h];
}
}
}
}


model {    
    sigma ~ uniform(0,2);
    aOrder ~ normal(0,2);
    for (d in 1:NxPredictors) {
        mu[d] ~ normal(0, 10);
        for (l in 1:NxSubjects){
          beta[l,d] ~ normal(mu[d],sigma);
          }
        }
    for (k in 1:NxSubjects){
        
        
        // Neutral condition choices
        ChoiceA[k,:] ~ bernoulli(inv_logit(PredictorsA[k,:]*beta1[k,:]));

        ChoiceB[k,:] ~ bernoulli(inv_logit(PredictorsB[k,:]*beta2[k,:]));
     
          }    
      
    }
    

    
'''
data_input = {
              'ChoiceA':switching,#choice data,
              'ChoiceB':switching1a,#choice data
              'NxChoices':140,
              'NxSubjects':17,
              'NxPredictors':7,
              'PredictorsA':predictors,
              'PredictorsB':predictors1a
    
}

controls={'adapt_delta':0.9,
         'max_treedepth':18}

model_fit_Anova = StanModel(model_code=model_input)
fit_model_all = model_fit_Anova.sampling(data=data_input,iter=2000,control=controls)
```

# SF + RR+ MF + H  model

i=state, j=feature, t=trial

### SF = successor feature learning

Update chosen state:

$\text{feature}_{i,j,t+1}=\text{feature}_{i,j,t} + \alpha_{chosen}(\text{O}_{feature,t}-\text{feature}_{i,j,t})$

Update counterfactual feedback:

$\text{feature}_{i,j,t+1}=\text{feature}_{i,j,t} + \alpha_{countefactual}(\text{O}_{feature,t}-\text{feature}_{i,j,t})$

$QSF =\begin{bmatrix}
\text{feature}_{1,1} & \text{feature}_{1,2}  \\
\text{feature}_{2,1} & \text{feature}_{2,2}  \\
\end{bmatrix} \begin{bmatrix}
\text{weight}_1\\
\text{weight}_2\\
\end{bmatrix}$

$\vec{Attention} = \begin{bmatrix}
\beta_{reward}\\
\beta_{predator}\\
\end{bmatrix}$

$\vec{\text{Instructed Reward Function}} =
\begin{bmatrix}
1\\
0\\
\end{bmatrix}$ or
$\begin{bmatrix}
0\\
-1\\
\end{bmatrix}$

$\text{weight}=\vec{\text{Instructed Reward Function}} \odot \vec{\text{Attention}}$

### Resource rational attentional shifting

In reward-rich environment (80 trials, counterbalanced):

$\vec{Attention} = \begin{bmatrix}
\beta_{reward} + \beta_{change,reward}\\
\beta_{predator}+ \beta_{change,predator}\\
\end{bmatrix}$

In predator-rich environment (80 trials, counterbalanced):

$\vec{Attention} = \begin{bmatrix}
\beta_{reward} - \beta_{change,reward}\\
\beta_{predator} - \beta_{change,predator}\\
\end{bmatrix}$


### MF = Model-free learning

$\text{QMF}_{i,j,t+1}=\text{QMF}_{i,j,t} + \alpha_{SF}(\text{O}_{reward,t}-\text{QMF}_{i,j,t})$

### H = Heuristic Strategy

$QH=\begin{bmatrix}
\text{feature}_{1,1} & \text{feature}_{1,2}  \\
\text{feature}_{2,1} & \text{feature}_{2,2}  \\
\end{bmatrix} \begin{bmatrix}
\text{reward}\\
\text{predator}\\
\end{bmatrix}$

$\text{reward} \sim \mathcal{N}(\mu,\sigma)$

$\text{predator} \sim \mathcal{N}(\mu,\sigma)$

### Stochastic choice process

$\forall{i}, Q_i = \beta_{MF}QMF_i + QSF_i + QH_i$

$p(Choice=1) = \frac{e^{Q_i}}{\sum_{k}{e^{Q_k}}}$
